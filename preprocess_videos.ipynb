```python

# Importing all necessary libraries
import cv2
import os
from os import listdir
from os.path import isfile, join
  
# Read the video from specified path
def extract_frames(folder,cat):
   
    
    try:
        
        # creating a folder named data
        if not os.path.exists('Frames/'+cat+'/'+folder):
            os.makedirs('Frames/'+cat+'/'+folder)
    
    # if not created then raise error
    except OSError:
        print ('Error: Creating directory of '.format(cat+'/'+folder))
   
    input_dir='/home/user/Jeba/CAREADD_PLAYvideos/recordings/Annotated_New/TD group/'

    #video_files_path = [join(input_dir+folder, f) for f in listdir(input_dir+folder) ]
    video_file_names = [f for f in listdir(input_dir+folder) ]
    frame_skip=1200
    i=0
    #print(video_files)
    frame_count=0
    for video_file in video_file_names:
        cam = cv2.VideoCapture(join(input_dir+folder,video_file))
        while(True):
            
            # reading from frame
            ret,frame = cam.read()
            #print(ret)
            if ret:
                # if video is still left continue creating images
                
                if i > frame_skip - 1:
                    frame_count += 1
                    #print(video_file+'\n'+video_file[:-3])
                    name = 'Frames/'+cat+'/'+folder +'/'+ folder+'_'+video_file[:-4] +'_frame_'+ str(frame_count)+'.jpg'
                    cv2.imwrite(name, frame)
                    i = 0
                    continue
                i += 1
                
            else:
                break
    
    # Release all space and windows once done
        cam.release()
        cv2.destroyAllWindows()
```

```python
list_of_DD_videos=['21_01_25_01_DD','21_02_03_01_DD','21_02_20_01_DD','21_10_13_01_DD','21_11_10_01_DD']
for item in list_of_DD_videos:
    extract_frames(item,'DD')
```
